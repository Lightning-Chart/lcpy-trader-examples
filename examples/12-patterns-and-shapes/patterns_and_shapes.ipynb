```python
from lightningchart_trader import TAChart
from datetime import datetime

LICENSE_KEY_PATH = "../../license_key.txt"
DATA_PATH = "../../data/Apple Inc. (AAPL) - 10y.csv"

# Load the license key
license_key = open(LICENSE_KEY_PATH).read()

chart = TAChart(license_key=license_key)

# Set chart title and theme
chart.set_chart_title('Cycle Lines, Sine Wave and Elliot Wave Patterns')
chart.set_color_theme('cyberSpace')

# Load CSV data
chart.load_csv(DATA_PATH)

# Adding Cycle Lines
cycle_lines = chart.add_cycle_lines(83, 50.7, 105, 41.6)
cycle_lines.set_time_zone_count(5)
cycle_lines.set_line_color('#FFFF00')  # Yellow color
cycle_lines.set_line_width(2)
cycle_lines.set_fill_enabled(False)

# Set time range
start_time = datetime(2018, 12, 31)
end_time = datetime(2019, 12, 30)
chart.set_time_range(start_time, end_time)

# Adding Sine Wave
sine_wave = chart.add_sine_wave(83, 54.6, 105, 50.0)
sine_wave.set_magnetic(True)
sine_wave.set_smoothness(5)
sine_wave.set_line_color('#F087FF')  # Light pink color
sine_wave.set_line_width(2)

# Adding Elliot Wave
elliot_wave = chart.add_elliot_wave(
    waveType=0,
    startX=2,
    startY=34.0,
    secondX=12,
    secondY=37.7,
    thirdX=16,
    thirdY=36.3,
    fourthX=25,
    fourthY=41.9,
    fifthX=27,
    fifthY=40.4,
    sixthX=42,
    sixthY=42.6,
    seventhX=46,
    seventhY=40.7,
    eighthX=56,
    eighthY=47.4,
    ninthX=58,
    ninthY=44.1,
)
elliot_wave.set_line_color('#00FF00')  # Green color
elliot_wave.set_line_width(2)
elliot_wave.set_wave_type(0)

# # Open Chart in notebook
chart.open()
```
